```python
import pandas as pd
import numpy as np
import os
import requests
from tqdm import tqdm
```

```python
df = pd.read_csv('fitzpatrick17k.csv')
df
```

```python
df['url'].dropna()
```

```python
for i in tqdm(df['url'].dropna()):
    name = i.split('/')[-1]
    path = os.path.join('/Users/antonkutsenko/Documents/Nova IMS current/DL', name)

    response = requests.get(i,headers={"User-Agent": "XY"})
    if response.status_code == 200:
        with open(path, 'wb') as f:
            f.write(response.content)
    else:
        print(f"Failed to download image from '{i}'. Status code: {response.status_code}")
```
